```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
#Cargamos el dataset
ventas = pd.read_csv("ventas-por-factura.csv")
```

```python
ventas
```

```python
#Mostramos las primeras filas 
ventas.head()
```

```python
#Mostramos las ultimas filas 
ventas.tail()
```

```python
#Vemos la cantidad de filas y columnas
ventas.shape
```

```python
#Revisamos información general del dataset
ventas.info()
```

```python
ventas.describe()
```

```python
#Creamos una copia
ventas_copia = ventas.copy()
```

```python
#Valores nulos por columna
ventas.isnull().sum()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(ventas.isnull(), cbar=False)
plt.title("Mapa de valores nulos")
plt.show()
```

```python
#Contamos registros duplicados
ventas.duplicated().sum()
```

```python
#Borramos datos duplicados
ventas.drop_duplicates(inplace=True)
```

```python
ventas.duplicated().sum()
```

```python
#Revisamos el tipo de dato de "Monto"
ventas["Monto"].head()
```

```python
#Reemplazamos coma por punto decimal
ventas["Monto"] = ventas["Monto"].str.replace(",", ".")
```

```python
#Convertimos la columna a tipo float
ventas["Monto"] = ventas["Monto"].astype(float)
```

```python
ventas.info()
```

```python
#Convertimos la fecha al formato datetime
ventas["Fecha de factura"] = pd.to_datetime(ventas["Fecha de factura"], errors="coerce")
```

```python
ventas.info()
```

```python
#Identificamos las facturas canceladas
facturas_canceladas = ventas[ventas["N° de factura"].astype(str).str.startswith("C")]
```

```python
facturas_canceladas.head()
```

```python
#Contamos las facturas canceladas 
facturas_canceladas.shape
```

```python
#Eliminamos las facturas canceladas
ventas = ventas[~ventas["N° de factura"].astype(str).str.startswith("C")]
```

```python
facturas_canceladas = ventas[ventas["N° de factura"].astype(str).str.startswith("C")]
```

```python
facturas_canceladas.head()
```

```python
#Buscamos cantidades negativas
ventas[ventas["Cantidad"] < 0]
```

```python
#Buscamos montos negativos
ventas[ventas["Monto"] < 0]
```

```python
#Reemplazamos las cantidades negativas por nulos
ventas.loc[ventas["Cantidad"] < 0, "Cantidad"] = np.nan
```

```python
#Reemplazamos los montos negativos por nulos
ventas.loc[ventas["Monto"] < 0, "Monto"] = np.nan
```

```python
#Calculamos la mediana de cantidad
mediana_cantidad = ventas["Cantidad"].median()
```

```python
mediana_cantidad
```

```python
#Calculamos la mediana de monto
mediana_monto = ventas["Monto"].median()
```

```python
mediana_monto
```

```python
#Reemplazamos nulos en cantidad
ventas["Cantidad"] = ventas["Cantidad"].fillna(mediana_cantidad)
```

```python
#Reemplazamos nulos en monto
ventas["Monto"] = ventas["Monto"].fillna(mediana_monto)
```

```python
ventas.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(ventas["Monto"], bins=30, kde=True)
plt.title("Distribución del monto")
plt.xlabel("Monto")
plt.ylabel("Frecuencia")
plt.show()
```

```python
#Agrupamos ventas por país
ventas_pais = ventas.groupby("País")["Monto"].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,6))
ventas_pais.plot(kind="bar")
plt.title("Ventas por país")
plt.xlabel("País")
plt.ylabel("Monto total")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=ventas["Monto"])
plt.title("Detección de outliers en monto")
plt.show()
```

```python
#Exportamos dataset limpio
ventas.to_csv("Ventas-por-factura-limpio.csv", index=False)
```
