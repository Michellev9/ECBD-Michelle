```python
cantidad = 1000000
```

```python
import random
```

```python
influencers = ["AnaTravel", "CarlosBeach", "SofiaLifestyle", "LuisTrips", "MarthaSun", "DiegoExplorer"],
tipos_post = ["Foto", "Reel", "Historia", "Video", "Foto", "Reel"]
```

```python
datos = {
    "influencer": [random.choice(influencers) for _ in range(cantidad)],
    "likes": [random.randint(100, 50000) for _ in range(cantidad)],
    "comentarios": [random.randint(10, 5000) for _ in range(cantidad)],
    "fecha": pd.date_range(start="2025-01-01", periods=cantidad, freq="min"),
    "tipo_post": [random.choice(tipos_post) for _ in range(cantidad)],
}
```

```python
df_grande = pd.DataFrame(datos)
```
